### 1.Plot the GDP as a function of the year in Jupyter as shown below.

```python
# import numpyï¼Œmatplotlib and math

import numpy as np
import matplotlib.pyplot as plt
```

```python
# input data

x1=np.linspace(1930,2010,9)
y1=[1.015,1.33,2.29,3.26,4.951,6.759,9.366,13.131,15.599]
```

```python
#Plot figure 1

plt.figure(1) #Generate figure
plt.plot(x1,y1,"r--") #Plot a straight line graph
plt.scatter(x1,y1,c='r',marker='.',linewidths=3) #Point the data points
plt.xlabel("Year") #Set the title of the x axis
plt.ylabel("Trillions of $") #Set the title of the y axis
plt.title("Real US GDP (in trillions)") #Set the title of the figure
```

       Figure 1:Real US GDP as a function of the year.

### 2.Based on the obtained plot,what is a possible mathematical relationship between GDP and Year?Carefully argue this point and see if you can lear to discuss this point within a markdown cell of Jupyter using mathematical notation.

The increase of GDP(y) is probably an exponential function of Year(x).Assuming that the increase rate keeps a constant r in this period,the mathematical relationship possibly is:
$$y_{i+1}=x^ry_i$$
As a result,determine r by:
$$r=(\frac{y_{2010}}{y_{1930}})^{\frac{1}{2010-1930}}=1.0308246274290083$$
So the expected mathematical relationship between y and x is:
$$y=x^{1.0308246274290083}$$
Plot the curve:

```python
#define x2 and y2

r1=pow(15.599/1.015,1/90) # Increase rate
x2=np.linspace(1930,2010,91)
y2=[1.015]
i=0
while i <= 89:
    y2.append(y2[i]*r1)
    i+=1
```

```python
plt.figure(2) #Generate figure
plt.plot(x1,y1,"r--") #Previous plot obtained (actual GDP data)
plt.plot(x2,y2,"b--") #New plot (the obtained mathematical relationship)
plt.scatter(x1,y1,c='r',marker='.',linewidths=3) #Point the data points
plt.xlabel("Year") #Set the title of the x axis
plt.ylabel("Trillions of $") #Set the title of the y axis
plt.title("Real US GDP (in trillions)") #Set the title of the figure
```

Figure 2:Real US GDP (red) & Obtained mathematical relationship (blue)

The new plot doesn't fit the previous plot well.Taking the subprime crisis started in 2007 into consideration,the point (2010,15.599) might be a outlier.So abandon the point and re-determine r with (1930,1.015) and (2000,13.131):
$$r=(\frac{y_{2000}}{y_{1930}})^{\frac{1}{2000-1930}}=1.0325186311935777$$
So the corrected mathematical relationship between y and x is:
$$y=x^{1.0325186311935777}$$
Plot the curve:

```python
#define y3

r2=pow(13.131/1.015,1/80) # Increase rate
x2=np.linspace(1930,2010,91)
y3=[1.015]
i=0
while i <= 89:
    y3.append(y3[i]*r2)
    i+=1
```

```python
plt.figure(3) #Generate figure
plt.plot(x1,y1,"r--") #Previous plot obtained (actual GDP data)
plt.plot(x2,y2,"b--") #Uncorrected plot (mathematical relationship)
plt.plot(x2,y3,"y--") #corrected plot (corrected mathematical relationship)
plt.scatter(x1,y1,c='r',marker='.',linewidths=3) #Point the data points
plt.xlabel("Year") #Set the title of the x axis
plt.ylabel("Trillions of $") #Set the title of the y axis
plt.title("Real US GDP (in trillions)") #Set the title of the figure
```

Figure 3:Real US GDP (red),Uncorrected mathematical relationship (blue),
         corrected mathematical relationship(yellow).

The corrected plot fits the plot of actual GDP data between 1930 and 2000,abandoning (2010,15.599).
