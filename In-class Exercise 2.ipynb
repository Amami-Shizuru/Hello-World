```python
import numpy as np
import math
from numpy import linalg as la
import matplotlib.pyplot as plt
import random
```

```python
Lambda1=np.mat([[4,0],[0,1]])
Lambda2=np.mat([[1,0],[0,5]])
U=np.mat([[1/math.sqrt(2),-1/math.sqrt(2)],[1/math.sqrt(2),1/math.sqrt(2)]])
Sigma1=U*Lambda1*U.T
Sigma2=U*Lambda2*U.T
Sigma3=np.eye(2)
mu1=[1,0]
mu2=[-1,0]
mu3=[0,-10]
X1=[[0 for i in range(2)] for j in range(1000)]
X2=[[0 for i in range(2)] for j in range(1000)]
X3=[[0 for i in range(2)] for j in range(1000)]
for i in range(1000):
    X1[i]=np.random.multivariate_normal(mu1, Sigma1)    
    X2[i]=np.random.multivariate_normal(mu2, Sigma2)    
    X3[i]=np.random.multivariate_normal(mu3, Sigma3)
X1=np.array(X1)
X2=np.array(X2)
X3=np.array(X3)
plt.scatter(X1[:,0],X1[:,1],color='r')
plt.scatter(X2[:,0],X2[:,1],color='g')
plt.scatter(X3[:,0],X3[:,1],color='b')
```

```python
#initialization
X1new=[[0 for i in range(2)] for j in range(1)]
X2new=[[0 for i in range(2)] for j in range(1)]
X3new=[[0 for i in range(2)] for j in range(1)]
for i in range(1000):
    x=random.randint(1,3)
    y=random.randint(1,3)
    z=random.randint(1,3)
    if x==1:
        X1new.append(X1[i])
    elif x==2:
        X2new.append(X1[i])
    else:
        X3new.append(X1[i])
        
    if y==1:
        X1new.append(X2[i])
    elif y==2:
        X2new.append(X2[i])
    else:
        X3new.append(X2[i])
        
    if z==1:
        X1new.append(X3[i])
    elif z==2:
        X2new.append(X3[i])
    else:
        X3new.append(X3[i])
        
del X1new[0]
del X2new[0]
del X3new[0]
X1new=np.array(X1new)
X2new=np.array(X2new)
X3new=np.array(X3new)
plt.scatter(X1new[:,0],X1new[:,1],color='r')
plt.scatter(X2new[:,0],X2new[:,1],color='g')
plt.scatter(X3new[:,0],X3new[:,1],color='b')
```

```python
X1=X1new
X2=X2new
X3=X3new
t=0

while(1):
    mu1=np.mean(X1,axis=0)
    mu2=np.mean(X2,axis=0)
    mu3=np.mean(X3,axis=0)
    X1new=[[0 for i in range(2)] for j in range(1)]
    X2new=[[0 for i in range(2)] for j in range(1)]
    X3new=[[0 for i in range(2)] for j in range(1)]
    for i in range(len(X1)):
        d11=np.linalg.norm(X1[i]-mu1)
        d12=np.linalg.norm(X1[i]-mu2)
        d13=np.linalg.norm(X1[i]-mu3)
        if min(d11,d12,d13)==d11:
            X1new.append(X1[i])
        elif min(d11,d12,d13)==d12:
            X2new.append(X1[i])
        else:
            X3new.append(X1[i])
        
    for i in range(len(X2)):
        d21=np.linalg.norm(X2[i]-mu1)
        d22=np.linalg.norm(X2[i]-mu2)
        d23=np.linalg.norm(X2[i]-mu3)
        if min(d21,d22,d23)==d21:
            X1new.append(X2[i])
        elif min(d21,d22,d23)==d22:
            X2new.append(X2[i])
        else:
            X3new.append(X2[i])
        
    for i in range(len(X3)):
        d31=np.linalg.norm(X3[i]-mu1)
        d32=np.linalg.norm(X3[i]-mu2)
        d33=np.linalg.norm(X3[i]-mu3)
        if min(d31,d32,d33)==d31:
            X1new.append(X3[i])
        elif min(d31,d32,d33)==d32:
            X2new.append(X3[i])
        else:
            X3new.append(X3[i])

    del X1new[0]
    del X2new[0]
    del X3new[0]
    t+=1
    if len(X1)==len(X1new) and len(X2)==len(X2new) and len(X3)==len(X3new):
        if X1==X1new and X2==X2new and X3==X3new:
            break
        else:
            X1=X1new
            X2=X2new
            X3=X3new
    else:
        X1=X1new
        X2=X2new
        X3=X3new
        
X1=np.array(X1)
X2=np.array(X2)
X3=np.array(X3)
plt.scatter(X1[:,0],X1[:,1],color='r')
plt.scatter(X2[:,0],X2[:,1],color='g')
plt.scatter(X3[:,0],X3[:,1],color='b')
print("t=",t)
```
